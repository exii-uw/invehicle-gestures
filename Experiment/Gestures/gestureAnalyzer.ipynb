```python
def create_info():
    info = {}
    for pid in pid_list:
        info[pid] = {}
        for mode in modes:
            info[pid][mode] = [[[], 0, [], 0, [], 0] for i in range(blocks+1)]
    return info
```

```python
def find_air_entity(air_lst, prompt_finish):
    difference = 9999999999999999
    for line in air_lst:
        current_difference = int(line[2]) - int(prompt_finish) + offset
        if current_difference > 0 and current_difference < difference:
            difference = current_difference
            entity = line
    return entity
```

```python
def find_log_entity(air_log, timestamp):
    for entity in air_log:
        if int(entity['TIME']['F']) == timestamp:
            return entity
#     print(timestamp)
    return None
```

```python
def get_slope(entity):
    import math
    import numpy as np
    diffX, diffY = int(entity['COR_X']['L']), int(entity['COR_Y']['L'])
    if diffX == 0:
        if diffY > 0:
            return [math.inf, [np.sign(diffX), np.sign(diffY)]]
        else:
            return [-math.inf, [np.sign(diffX), np.sign(diffY)]]
    slope = diffY/diffX
    return [round(slope, 4), [np.sign(diffX), np.sign(diffY)]]
```

```python
def getNeighbors(direct):
    directs = ['Up Left', 'Up', 'Up Right', 'Right', 'Down Right', 'Down', 'Down Left', 'Left', 'Up Left', 'Up']
    index = directs.index(direct, 1)
    return [directs[index-1], directs[index], directs[index+1]]
getNeighbors('Up Left')
```

```python
def check_value(pid, mode, info, metric, increment, block_length, reported, expected, approach):
    block = info[pid][mode][-1][metric] // block_length
#     print(info[pid][mode][-1][metric], block)
    if metric == 1:
        if (approach in ['Normal', 'D4'] and expected == reported) or \
        (approach == 'AN' and reported in getNeighbors(expected)): # if correct for accuracy
            info[pid][mode][block][metric-1].append(increment)
        else:
            info[pid][mode][block][metric-1].append(0)
    else:
        if increment >= 0: # if correct for time values
            info[pid][mode][block][metric-1].append(increment)
    info[pid][mode][-1][metric] += 1
    info[pid][mode][block][metric] += 1
    return
```

```python
def get_sum_count(offset, approach):
    j, c = 0, 0
    rows, directions = [], []
    discard = {'Air': 0, 'Swipe': 0, 'Tap': 0}
    for i in range(len(data_srf_list)):
        line = data_srf_list[i]
        pid, mode, prompt_dir, prompt_start, prompt_finish, perform_dir, perform_start, perform_finish = int(line[0]), line[1][:-9], line[2], line[3], line[4], line[5], line[6], line[7]
        # add air data if necessary
        if mode == 'Air':
            entity = data_air_list[j]
            air_pid, perform_dir, perform_start, perform_finish = int(entity[0]), entity[1], int(entity[2])+offset, int(entity[3])+offset
            log_entity = find_log_entity(air_log, perform_finish-offset)
            if log_entity != None:
                directions.append([pid, prompt_dir] + get_slope(log_entity))
            else:
#                 print(pid, air_pid, perform_finish-offset)
                print('Air Unperformed by', pid)
            if air_pid != pid:
                print(line)
                print(entity)
                print('____________________________________________')
            j += 1
        block_length = 32//blocks
        if approach == 'D4': # Discard 4
            if prompt_dir not in ['Up', 'Right', 'Left', 'Down']:
                continue
            block_length = 4
        block = c//16 + 1
        c += 1
        if c == 32:
            c = 0
        response_time = int(perform_start) - int(prompt_finish)
        completion_time = int(perform_finish) - int(perform_start)
        if (mode != 'Air' and int(perform_start) == 0) or (mode == 'Air' and int(perform_start) == offset):
            response_time, completion_time = 7500, 0
        if mode == 'Air':
            if pid == 8:
                response_time += 15000
            elif pid in [2, 18]:
                response_time += 250
        time = response_time + completion_time
        if response_time > 0:
            rows.append([pid, mode, block, prompt_dir, perform_dir, response_time, completion_time, time])
        else:
            print(pid, mode, response_time)
            discard[mode] += 1
#         check_value(pid, mode, info, 1, 1, block_length, perform_dir, prompt_dir, approach) # accuracy
#         check_value(pid, mode, info, 3, int(perform_start) - int(prompt_finish), block_length, 0, 0, approach) # response time
#         check_value(pid, mode, info, 5, int(perform_finish) - int(perform_start), block_length, 0, 0, approach) # completion time
#         print(info[pid][mode])
    info_df = pd.DataFrame(rows, columns = ['PID', 'Method', 'Block', 'Prompt Dir', 'Perform Dir', 'Response Time', 'Completion Time', 'Selection Time'])
    print(discard)
    return info_df, directions
```

```python
def get_error(line):
    from math import sqrt
    accuracy = 100*line[0]/line[1]
    avg_response_time = int(line[2]/line[3])
    avg_completion_time = int(line[4]/line[5])
    return accuracy, avg_response_time, avg_completion_time
```

```python
def get_summary(info_df):
    rows = []
    for pid in pid_list:
        for mode in new_modes:
            for block in range(blocks):
                df = info_df[(info_df['Block'] == block+1) & (info_df['Method'] == mode) & (info_df['PID'] == pid)].copy()
                avg_response_time = df['Response Time'].mean()
                avg_completion_time = df['Completion Time'].mean()
                avg_time = df['Selection Time'].mean()
                accuracy = df['Accuracy'].mean()
                rows.append([pid, mode, block+1, accuracy, avg_response_time, avg_completion_time, avg_time])
    return pd.DataFrame(rows, columns = ['PID', 'Method', 'Block', 'Accuracy', 'Response Time', 'Completion Time', 'Selection Time'])
```

```python
def remove_outliers(org_df, column):
    import numpy as np
    data = org_df.loc[:,column]
    df = org_df.copy()
    # std
    datalist = data.tolist()
    import statistics
    mu, std = statistics.mean(datalist), statistics.stdev(datalist)
    max_val = mu + 3*std
    min_val = mu - 3*std
    print(int(min_val), int(mu), int(max_val))
    # quarters
#     q75, q25 = np.percentile(data,[75,25])
#     intr_qr = q75-q25
#     max_val = q75+(1.5*intr_qr)
#     min_val = q25-(1.5*intr_qr)
    # filter
    df.loc[org_df[column] < min_val, column] = 'OL'
    df.loc[org_df[column] > max_val, column] = 'OL'
    return df[df[column] != 'OL']
```

```python
def plot_and_save(df, x_parameter, y_parameter, x_label, y_label, title, isGrouped, hue):
    import pandas as pd
    import seaborn as sns
    import matplotlib.pyplot as plt
    
    sns.set_theme(style="whitegrid")
    sns.set(rc = {'figure.figsize':(15,8)})

    fig, ax = plt.subplots(figsize=(15,8))
    if isGrouped:
        ax = sns.barplot(x=x_parameter, y=y_parameter, data=df, capsize=.2, hue=hue)
    else:
        ax = sns.barplot(x=x_parameter, y=y_parameter, data=df, capsize=.2)
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    fig.savefig('Figs/'+title+".png")
    
    plt.show()
```

```python
def factorize_xticks(ax):
    labels = []
    for label in ax.xaxis.get_ticklabels():
        # label.set_text(wrap_factor(label.get_text()))
        # label.set_usetex(True)
        label.set_text(label.get_text().upper())
        labels += [label]
    ax.xaxis.set_ticklabels(labels)
```

```python
def cute_plot(df, x_label, x_label_name, x_order, y_label, y_label_name, y_bounds, title, duality, legend_position, fontsize, frmt='pdf'):
    import seaborn as sns
    import matplotlib
    import matplotlib.pyplot as plt
    reg_palette = sns.color_palette("colorblind", 5)
    sns.set_palette(reg_palette)
    light_palette = [sns.set_hls_values(x, l=0.65) for x in reg_palette]
    sns.set_style("whitegrid")
    sns.set(font_scale = 1.2)
    
    fig = plt.figure()
    fig.set_figwidth(6)
    fig.set_figheight(4)
    gs = fig.add_gridspec(ncols=1, nrows=1, wspace=0.1)
    ax = fig.add_subplot(gs[0, 0])
    patches = []
    if duality:
        for metric, palette in [("Selection Time", reg_palette), ("Response Time", light_palette)]:
            ax = sns.barplot(data=df, x=x_label, order=x_order, y=metric, label=metric, ci=95, ax=ax, n_boot=10000, \
                palette=palette,)
            for i,t in enumerate(t.get_text() for t in ax.get_xticklabels()):
                patches.append(matplotlib.patches.Patch(color=palette[i], label= metric + ' ' + t.upper()))
    else:
        metric, palette = y_label, reg_palette
        ax = sns.barplot(data=df, x=x_label, order=x_order, y=metric, label=metric, ci=95, ax=ax, n_boot=10000, \
                palette=palette,)
        for i,t in enumerate(t.get_text() for t in ax.get_xticklabels()):
                patches.append(matplotlib.patches.Patch(color=palette[i], label= metric + ' ' + t.upper()))
    ax.set_ylim(ymin=y_bounds[0], ymax=y_bounds[1])
    ax.set_xlabel(x_label_name, fontsize=fontsize)
#     ax.set_xlabel(x_label_name.upper())
    factorize_xticks(ax)
#     ax.set_ylabel(y_label_name)
    ax.set_ylabel(y_label_name, fontsize=fontsize+1)
#     if duality:
#         ax.legend(handles=patches, loc=legend_position, fontsize=fontsize-4) # set legend
    # Remove spines
    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.xaxis.set_ticklabels(ax.xaxis.get_ticklabels(), fontsize=fontsize) # Reduce tick label size for x axis
#     ax.yaxis.set_ticklabels(fontsize=fontsize) # Reduce tick label size for x axis
    fig.savefig('Figs/'+title+'.'+frmt, dpi=300, bbox_inches = "tight")
    plt.show()
```

```python
def get_x_vals(slope, signs):
    xd, yd = signs[0], signs[1]
    x = 1/math.sqrt(pow(slope, 2) + 1)
    x_right, x_left = 0, 0
    if xd == 1:
        x_right = x
    elif xd == -1:
        x_left = -x
    x_vals = np.array([x_left, x_right])
    y_vals = slope * x_vals
    return x_vals, y_vals
```

```python
def is_between(direction, slope, signs, step):
    xd, yd = signs[0], signs[1]
    if direction == 'Up':
        return yd == 1 and (slope <= math.tan(math.pi/2+step) or slope >= math.tan(math.pi/2-step))
    elif direction == 'Down':
        return yd == -1 and (slope <= math.tan(math.pi/2+step) or slope >= math.tan(math.pi/2-step))
    elif direction == 'Right':
        return xd == 1 and (slope <= math.tan(step) and slope >= math.tan(2*math.pi-step))
    elif direction == 'Left':
        return xd == -1 and (slope <= math.tan(step) and slope >= math.tan(2*math.pi-step))
    elif direction == 'Up Right':
        return xd == 1 and yd == 1 and (slope >= math.tan(step) and slope <= math.tan(math.pi/2-step))
    elif direction == 'Down Left':
        return xd == -1 and yd == -1 and (slope >= math.tan(step) and slope <= math.tan(math.pi/2-step))
    elif direction == 'Up Left':
        return xd == -1 and yd == 1 and (slope <= math.tan(math.pi-step) and slope >= math.tan(math.pi/2+step))
    elif direction == 'Down Right':
        return xd == 1 and yd == -1 and (slope <= math.tan(math.pi-step) and slope >= math.tan(math.pi/2+step))
```

```python
def plot_directions(direction, limit_slope, signs, dirs):
    import numpy as np 
    import matplotlib.pyplot as plt 

    title = direction + ' Performance Distribution'
    left_slope, right_slope = limit_slope[0], limit_slope[1]
    left_signs, right_signs = signs[0], signs[1]
    angle = np.linspace( 0 , 2 * np.pi , 150 ) 
    radius = 1
    x = radius * np.cos( angle ) 
    y = radius * np.sin( angle ) 

    figure, axes = plt.subplots( 1 ) 
    axes.plot( x, y ) 
    axes.set_aspect( 1 )
    
    # plot axis lines
    plt.plot([0, 1], [0, 0], '--', color='blue', linewidth=1.5)
    plt.plot([0, -1], [0, 0], '--', color='blue', linewidth=1.5)
    plt.plot([0, 0], [0, 1], '--', color='blue', linewidth=1.5)
    plt.plot([0, 0], [0, -1], '--', color='blue', linewidth=1.5)
    
    # plot performed lines
    for line in dirs:
        slope, signs_list = line[0], line[1]
        x_vals, y_vals = get_x_vals(slope, signs_list)
        if is_between(direction, slope, signs_list):
            plt.plot(x_vals, y_vals, '-', color='green')
        else:
            plt.plot(x_vals, y_vals, '-', color='red')

    # plot limit lines
    x_vals, y_vals = get_x_vals(left_slope, left_signs)
    plt.plot(x_vals, y_vals, '--', color='yellow', linewidth=4)
    x_vals, y_vals = get_x_vals(right_slope, right_signs)
    plt.plot(x_vals, y_vals, '--', color='yellow', linewidth=4)
    
    plt.axis('off')
    plt.title(title)
    plt.show()
    figure.savefig('Figs/'+title+'.pdf', dpi=300)
```

```python
def remove_outliers(org_df, column):
    import numpy as np
    data = org_df.loc[:,column]
    df = org_df.copy()
    # std
    datalist = data.tolist()
    import statistics
    mu, std = statistics.mean(datalist), statistics.stdev(datalist)
    max_val = mu + 3*std
    min_val = mu - 3*std
    # quarters
#     q75, q25 = np.percentile(data,[75,25])
#     intr_qr = q75-q25
#     max_val = q75+(1.5*intr_qr)
#     min_val = q25-(1.5*intr_qr)
    # filter
    df.loc[org_df[column] < min_val, column] = 'OL'
    df.loc[org_df[column] > max_val, column] = 'OL'
    return df[df[column] != 'OL']
```
