```python
import pandas as pd
data = pd.read_excel('Participant Data (Responses).xlsx')
data
```

```python
column_names = list(data.columns)
translate_dict = {column_names[1]: 'PID', column_names[10]: 'Mental Demand', column_names[11]: 'Physical Demand', \
                  column_names[12]: 'Temporal Demand', column_names[13]: 'Performance', \
                  column_names[14]: 'Effort', column_names[15]: 'Frustration'}
odf = data.rename(translate_dict, axis = 'columns')
drop_list = [column_names[i] for i in [0, 2, 3, 4, 5, 6, 7, 8, 9]]
odf = odf.drop(drop_list, axis = 'columns')
odf
```

```python
rows = []
for index, row in odf.iterrows():
    row_air, row_tap, row_swipe = [row['PID'], 'Air'], [row['PID'], 'Tap'], [row['PID'], 'Swipe']
    for metric in list(odf.columns)[1:]:
        row_air.append(row[metric].split('/')[0])
        row_tap.append(row[metric].split('/')[1])
        row_swipe.append(row[metric].split('/')[2])
    rows.append(row_air)
    rows.append(row_tap)
    rows.append(row_swipe)
df = pd.DataFrame(rows, columns = ['PID', 'Condition'] + list(odf.columns)[1:])
df
```

```python
rows = []
for index, row in odf.iterrows():
    for metric in list(odf.columns)[1:]:
        rows.append([row['PID'], 'Air', metric, row[metric].split('/')[0]])
        rows.append([row['PID'], 'Tap', metric, row[metric].split('/')[1]])
        rows.append([row['PID'], 'Swipe', metric, row[metric].split('/')[2]])
df = pd.DataFrame(rows, columns = ['PID', 'Method', 'Metric', 'Score'])
df = df.replace('Air', 'MIDAIR')
df = df.replace('Tap', 'TAP')
df = df.replace('Swipe', 'SWIPE')
new_modes = ['TAP', 'SWIPE', 'MIDAIR']
df['Score'] = df['Score'].astype(int)
df['Method'] = pd.CategoricalIndex(df['Method'], ordered=True, categories=new_modes)
df = df.sort_values(by = ['PID', 'Method'])
df
```

```python
def plot_and_save(df, x_parameter, y_parameter, x_label, y_label, title, isGrouped, hue):
    import pandas as pd
    import seaborn as sns
    import matplotlib.pyplot as plt
    
    sns.set_theme(style="whitegrid")
    sns.set(rc = {'figure.figsize':(15,8)})
    sns.set(font_scale = 1.3)

    fig, ax = plt.subplots(figsize=(15,8))
    if isGrouped:
        ax = sns.barplot(x=x_parameter, y=y_parameter, data=df, capsize=.2, hue=hue)
    else:
        ax = sns.barplot(x=x_parameter, y=y_parameter, data=df, capsize=.2)
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    fig.savefig('Figs/'+title+".pdf", dpi = 300)
    
    plt.show()
```

```python
def factorize_xticks(ax):
    labels = []
    for label in ax.xaxis.get_ticklabels():
        # label.set_text(wrap_factor(label.get_text()))
        # label.set_usetex(True)
        label.set_text(label.get_text().upper())
        labels += [label]
    ax.xaxis.set_ticklabels(labels)
```

```python
def cute_plot(df, x_label, x_label_name, x_order, y_label, y_label_name, y_bounds, title, duality, legend_position, fontsize):
    import seaborn as sns
    import matplotlib
    import matplotlib.pyplot as plt
    reg_palette = sns.color_palette("colorblind", 7)
    sns.set_palette(reg_palette)
    light_palette = [sns.set_hls_values(x, l=0.65) for x in reg_palette]
    sns.set_style("whitegrid")
    sns.set(font_scale = 1.2)
    
    fig = plt.figure()
    fig.set_figwidth(15)
    fig.set_figheight(8)
    gs = fig.add_gridspec(ncols=1, nrows=1, wspace=0.1)
    ax = fig.add_subplot(gs[0, 0])
    patches = []
    metric, palette = y_label, reg_palette
    ax = sns.barplot(data=df, x=x_label, y=metric, hue='Method', ci=95, ax=ax, n_boot=10000, \
            palette=palette,)
#     for i,t in enumerate(t.get_text() for t in ax.get_xticklabels()):
#             patches.append(matplotlib.patches.Patch(color=palette[i], label= x_order[i] + ' ' + t.upper()))
    ax.set_ylim(ymin=y_bounds[0], ymax=y_bounds[1])
    ax.set_xlabel(x_label_name.upper(), fontsize=fontsize)
    factorize_xticks(ax)
    ax.set_ylabel(y_label_name, fontsize=fontsize+1)
#     if duality:
#         ax.legend(handles=patches, loc=legend_position) # set legend
    ax.legend(loc=legend_position) # set legend
    # Remove spines
    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.xaxis.set_ticklabels(ax.xaxis.get_ticklabels(), fontsize=fontsize) # Reduce tick label size for x axis
#     ax.yaxis.set_ticklabels(ax.yaxis.get_ticklabels(), fontsize=fontsize) # Reduce tick label size for x axis
    fig.savefig('Figs/'+title+".pdf", dpi = 300)
    plt.show()
```

```python
rs = []
for index, row in sdf.iterrows():
    remain = index % 3
    if remain == 1:
        r = [0, 0, 0]
    r[remain-1] = row['Score']
    if remain == 0:
        rs.append(r)
ndf = pd.DataFrame(rs, columns = ['TAP', 'SWIPE', 'MIDAIR'])
ndf
```

```python
# conduct the  Shapiro-Wilk Test
print(shapiro(sdf['Score']))
print(shapiro(sdf[sdf['Method'] == 'MIDAIR']['Score']))
print(shapiro(sdf[sdf['Method'] == 'TAP']['Score']))
print(shapiro(sdf[sdf['Method'] == 'SWIPE']['Score']))
```
